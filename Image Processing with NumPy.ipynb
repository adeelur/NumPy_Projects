# Image Processing with NumPy

This project involves basic image processing using NumPy, without external libraries like OpenCV or PIL.
- Load an image as a NumPy array
- Convert it to grayscale
- Apply image thresholding
- Implement edge detection using the Sobel filter

### Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
from imageio import imread
import matplotlib.image as mpimg
```

### Load Image as a NumPy Array

```python
img = 'C:/Users/adeel/Downloads/rooom.jpg'
room_image = mpimg.imread(img)
imgplot = plt.imshow(room_image)
plt.axis('off')
plt.show()
```

### Convert Image to Grayscale

```python
grayscale_image = room_image[:, :, 0] * 0.2989 + room_image[:, :, 1] * 0.587 + room_image[:,:,2] * 0.144

plt.imshow(grayscale_image, cmap='gray')
plt.axis('off')
plt.show()
```

### Convert Image to Negative of grayscale_image

```python
negative_image = 255 - grayscale_image

plt.imshow(negative_image, cmap='gray')
plt.axis('off')
plt.show()
```

### Apply Image Thresholding

```python
threshold_value = 128
binary_image = (grayscale_image > threshold_value) * 255
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
plt.show()
```

### Flipping Image

```python
flipped_horizontal = np.fliplr(room_image)
flipped_vertical = np.flipud(room_image)
fig, ax = plt.subplots(1,3, figsize=(15,5))

ax[0].imshow(room_image)
ax[0].axis('off')
ax[0].set_title('Normal Image')

ax[1].imshow(flipped_horizontal, cmap='gray')
ax[1].axis('off')
ax[1].set_title('Horizontal Flipped')

ax[2].imshow(flipped_vertical, cmap='gray')
ax[2].axis('off')
ax[2].set_title('Vertical Flipped')

plt.show()
```

### Implement Edge Detection (Sobel Filter)

```python
from scipy.ndimage import sobel

sobel_x = sobel(grayscale_image, axis=0)
sobel_y = sobel(grayscale_image, axis=1)
edge_image = np.hypot(sobel_x, sobel_y)

plt.imshow(edge_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
from scipy.ndimage import rotate

rotated_image = rotate(room_image, angle=45, mode='reflect')
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```
